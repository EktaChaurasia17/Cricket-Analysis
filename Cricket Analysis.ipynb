```python
player_of_match = df_match['player_of_match'].value_counts().head(20)
```

```python
ax = sns.barplot(x= player_of_match  , y= player_of_match.index );
plt.title("Which player had won most of the time as man of the match ?");
plt.xlabel('Count');
for bars in ax.containers:
    ax.bar_label(bars);
```

*This chart shows that who has won man of the match most of the time*

```python
df_match['result'].value_counts()
```

```python
winner = df_match['winner'].value_counts()
```

```python
ax = sns.barplot( x=winner , y= winner.index );
plt.xlabel('count');
plt.title('Which team had won many times?');
for bars in ax.containers:
    ax.bar_label(bars);
```

*This chart shows that which team has won many times*

```python
sns.barplot(y= df_match['toss_winner'],  x= df_match['toss_winner'].index , color='r' );
sns.barplot(y= df_match['winner'],  x= df_match['winner'].index );

```

*This chart representing the actual 'winner' verses 'toss_winner' . The dominating 'Red' color on some bars shows that this team has won the toss more times compare to the actaul winning the match.*


```python
played = pd.concat([df_match['team1'],df_match['team2']] , axis = 0)
```

```python
most_played = played.value_counts().head(20)
```

```python
most_played
```

```python
ax=sns.barplot(x=most_played , y= most_played.index)
plt.title('Which team had played many times?')
plt.xlabel('Count')
for bars in ax.containers:
    ax.bar_label(bars)
```

*This chart shows that which team had played many times*

```python
ax=df_match['toss_decision'].value_counts()
```

```python
ax
```

```python
plt.pie(ax , labels = ['Fielding ','Batting '], shadow = True , startangle = 90, autopct = '%1.1f%%');
plt.show()
```

*This chart shows that , the toss_winner team prefer to choose Fielding over Batting.*

```python
umpire = pd.concat([df_match['umpire1'],df_match['umpire2']], axis=0)
```

```python
top_umpires = umpire.value_counts(ascending=False).head(20)
```

```python
ax=sns.barplot(x= top_umpires , y = top_umpires.index);
plt.xlabel('Count');
plt.title('Most popular umpire');
for bars in ax.containers:
    ax.bar_label(bars);
```

*This chart shows that who is the very famous Umpire*
